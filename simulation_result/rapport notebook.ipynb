# Rapport de simulation SEIR

Nous avons rencontré différents cas après plusieurs execution 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Liste des fichiers à lire
fichiers = ["test1.txt", "test2.txt", "test3.txt", "test4.txt", "test5.txt", "test6.txt"]

# Création du graphique
fig, axes = plt.subplots(3, 2, figsize=(12, 8))
fig.suptitle("Graphiques pour les fichiers testi.txt")

for i, fichier in enumerate(fichiers):
    ax = axes[i // 2, i % 2]  # Sélectionne le sous-plot approprié

    # Lecture des données à partir du fichier
    df = pd.read_csv(fichier, sep=',', skipinitialspace=True)

    # Création des graphiques avec des lignes continues
    ax.plot(df['tour'], df['S'], label='S')
    ax.plot(df['tour'], df['E'], label='E')
    ax.plot(df['tour'], df['I'], label='I')
    ax.plot(df['tour'], df['R'], label='R')

    # Titre
    ax.set_title(f"Graphique pour {fichier}")
    ax.set_xlabel("Tour")
    ax.set_ylabel("Nombre")

# Légende commune pour les quatre graphiques
lines, labels = fig.axes[-1].get_legend_handles_labels()
fig.legend(lines, labels, loc='upper right')

# Ajustement de l'espacement entre les sous-plots
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Affichage des graphiques
plt.show()
```

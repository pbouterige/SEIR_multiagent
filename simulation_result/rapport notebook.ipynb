# Rapport de simulation SEIR

```python
import matplotlib.pyplot as plt

# Listes pour stocker les donnes
tours = []
S = []
E = []
I = []
R = []

# Lecture du fichier
with open("test.txt", "r") as fichier:
    lignes = fichier.readlines()
    for ligne in lignes[1:]:  # Commence  la deuxime ligne pour ignorer les en-ttes
        elements = ligne.strip().split(',')
        tour, s, e, i, r = map(int, elements)
        tours.append(tour)
        S.append(s)
        E.append(e)
        I.append(i)
        R.append(r)

# Cration du graphe
plt.figure(figsize=(10, 6))
plt.plot(tours, S, label='S')
plt.plot(tours, E, label='E')
plt.plot(tours, I, label='I')
plt.plot(tours, R, label='R')
plt.xlabel('Tour')
plt.ylabel('Nombre de personnes')
plt.title('volution de la population par tour')
plt.legend()
plt.grid(True)
plt.show()
```
